## Training & Evaluating Classification Models

### Logistic Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
log_model = LogisticRegression()
log_model.fit(X_train, y_train)

y_pred_log = log_model.predict(X_test)
```

```python
y_pred_log
```

```python
cm_log = confusion_matrix(y_test, y_pred_log)
sns.heatmap(cm_log, annot = True)
plt.show()
```

```python
print("Accuracy {} %".format( 100 * accuracy_score(y_test, y_pred_log)))
```

```python
print(classification_report(y_test, y_pred_log))
```

### RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
plt.show()
```

### ANN classifier

```python
import tensorflow as tf
```

```python
# Initialization of ANN
model = tf.keras.models.Sequential()
```

```python
# Adding of hidden layer and input layer
model.add(tf.keras.layers.Dense(units=500, activation='relu', input_shape=(50, )))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
```

```python
# Adding output layer - binary_classification - that's why sigmoid activation function
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# In case of non-binary classification - categorical_crossentropy
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 50)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```python
plt.plot(epochs_hist.history['accuracy'])
plt.title('Model Accuracy Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Accuracy')
plt.legend(['Training Accuracy'])
```

```python
# Testing Set Performance
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```
