## Reading all the similar files

```python
def concat_files(in_path, file_str, encoding, data_format, out_path, out_file_name):
    all_files = [i for i in glob.glob(file_str + '.' + data_format)]
    combined_data = pd.concat([pd.read_csv(f, encoding=encoding, low_memory = False) for f in all_files])
    
    return combined_data.to_csv(out_path + out_file_name + '.' + data_format, index=False)
```

## Merging all datasets

```python
def merge_dataset(df1, df2, keys, join_cond):
    df_final = df1.merge(df2, on = keys, how = join_cond)
    return df_final
```
