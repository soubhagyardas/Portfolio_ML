### Visualizing the dataset

```python
df_explore.duplicated().sum()
```

```python
df_rough = df_explore.copy()
```

```python
df_explore.compare(df_rough)
```

```python
df_rough_pd = df_explore.copy()
```

```python
df_rough_pd['MINIMUM_PAYMENTS'].fillna(df_rough_pd['MINIMUM_PAYMENTS'].mean(), inplace = True)
```

```python
df_rough_pd.isnull().sum()
```

```python
df_rough_pd['CREDIT_LIMIT'].fillna(df_rough_pd['CREDIT_LIMIT'].mean(), inplace = True)
```

```python
df_rough_pd.isnull().sum()
```

dropping custid

```python
df_rough_pd.drop("CUST_ID", axis = 1, inplace= True)
```

```python
plt.figure(figsize=(10,50))
for i in range(len(df_rough_pd.columns)):
  plt.subplot(17, 1, i+1)
  sns.distplot(df_rough_pd[df_rough_pd.columns[i]], kde_kws={"color": "b", "lw": 3, "label": "KDE"}, hist_kws={"color": "g"})
  plt.title(df_rough_pd.columns[i])

plt.tight_layout()
plt.show()
```

```python
f, ax = plt.subplots(figsize = (20,20))
sns.heatmap(df_rough_pd.corr(), annot = True)
plt.title('Correlation between the features')
plt.show()
```
