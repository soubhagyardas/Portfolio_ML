```python
def concat_files(in_path, file_str, encoding, data_format, out_path, out_file_name):
    all_files = [i for i in glob.glob(file_str + '.' + data_format)]
    combined_data = pd.concat([pd.read_csv(f, encoding='latin-1', low_memory = False) for f in all_files])
    
    return combined_data.to_csv(outgoing_path + out_file_name + '.' + data_format, index=False)
```
